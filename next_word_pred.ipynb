```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Import  Dataset

```python
df = pd.read_csv('qoute_dataset.csv')
```

Preprocessing Data

```python
quotes = df['quote']
```

```python
quotes = quotes.str.lower()

import string
quotes = quotes.str.translate(str.maketrans('', '', string.punctuation))
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
vocab_size = 10000

tokenizer = Tokenizer(num_words = vocab_size)

tokenizer.fit_on_texts(quotes)

print(tokenizer.word_index)
```

```python
sequence = tokenizer.texts_to_sequences(quotes)
```

```python
print(quotes[0])
print(sequence[0])
```

```python
X = []
y = []

for seq in sequence:
  for i in range(1, len(seq)):
    X.append(seq[:i])
    y.append(seq[i])
```

```python
max_len = max(len(s) for s in X)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
X_padded = pad_sequences(X, maxlen=max_len, padding='pre')
```

```python
y = np.array(y)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_one_hot = to_categorical(y, num_classes=vocab_size)
```

Creating Models

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,LSTM, Dense
```

```python
embedding_dim = 50
rnn_units = 128
```

```python
rnn_model = Sequential()

rnn_model.add(
    Embedding(
        input_dim=vocab_size,
        output_dim=embedding_dim,
        input_length=max_len
    )
)
rnn_model.add(
    SimpleRNN(
        units=rnn_units
    )
)
rnn_model.add(
    Dense(
        units=vocab_size,
        activation='softmax'
    )
)
```

```python
rnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
rnn_model.summary()
```

```python
lstm_model = Sequential()

lstm_model.add(
    Embedding(
        input_dim=vocab_size,
        output_dim=embedding_dim,
        input_length=max_len
    )
)
lstm_model.add(
    LSTM(
        units=rnn_units
    )
)
lstm_model.add(
    Dense(
        units=vocab_size,
        activation='softmax'
    )
)
```

```python
lstm_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
lstm_model.summary()
```

Training LSTM Model

```python
lstm_model.fit(
    X_padded,
    y_one_hot,
    epochs=100,
    batch_size=32,
    validation_split=0.1,
    verbose=1
)
```

```python
word_map = {}
for word, index in tokenizer.word_index.items():
    word_map[index] = word
```

```python
def predict_next_word(text, model, tokenizer, max_len, word_map):
    sequence = tokenizer.texts_to_sequences([text.strip().lower()])
    padded_sequence = pad_sequences(sequence, maxlen=max_len, padding='pre')
    prediction = model.predict(padded_sequence)
    return word_map[np.argmax(prediction)]
```

```python
text = "what is a"
predict_next_word(text, lstm_model, tokenizer, max_len, word_map)
```
